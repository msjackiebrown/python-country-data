# Country Data Visualization with Matplotlib

This notebook demonstrates how to visualize country area and population data using Matplotlib. We will create a scatter plot and two bar graphs to explore the relationships and distributions in the data.

## 1. Import Required Libraries

Let's import the libraries needed for data handling and plotting.

```python
import matplotlib.pyplot as plt

# Sample data for demonstration
country_names = ['Country A', 'Country B', 'Country C', 'Country D', 'Country E']
areas = [50000, 75000, 60000, 25000, 90000]
populations = [1000000, 2000000, 1500000, 500000, 3000000]

print('Country Names:', country_names)
print('Areas:', areas)
print('Populations:', populations)
```

<!-- Removed: Prepare Country Data section, as we now use hardcoded lists for sample data. -->

## 1. Bar Graph: Country Areas

Let's start by plotting a bar graph showing the area of each country. Bar charts are a simple and effective way to compare values across categories.

The following code creates a bar chart where each bar represents a country and its area:

```python
# Bar graph of country areas
#
# This code creates a bar chart that visually compares the area of each country.
# Each bar represents a country, and the height of the bar shows the area value.
# This makes it easy to see which countries are larger or smaller in area.
#
# Step-by-step explanation:
# 1. plt.figure(figsize=(10, 6))
#    - Creates a new figure for the plot with a width of 10 inches and height of 6 inches.
# 2. plt.bar(country_names, areas, color='orange')
#    - Draws a bar for each country, where the height of the bar is the area value. The bars are colored orange.
# 3. plt.title('Area by Country')
#    - Sets the title of the plot.
# 4. plt.xlabel('Country') and plt.ylabel('Area (sq km)')
#    - Label the x-axis and y-axis.
# 5. plt.xticks(rotation=45)
#    - Rotates the country names on the x-axis by 45 degrees for better readability.
# 6. plt.tight_layout()
#    - Adjusts the layout so labels and titles fit well.
# 7. plt.show()
#    - Displays the plot.

plt.figure(figsize=(10, 6))
plt.bar(country_names, areas, color='orange')
plt.title('Area by Country')
plt.xlabel('Country')
plt.ylabel('Area (sq km)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 2. Bar Graph: Country Populations

Next, let's plot a bar graph showing the population of each country. This helps us compare the population sizes visually.

**Explanation:**
- This code creates a bar chart where each bar represents a country, and the height of the bar shows the population value.
- The x-axis shows the country names, and the y-axis shows the population of each country.
- This makes it easy to see which countries have larger or smaller populations.

The following code creates a bar chart for populations:

```python
# Bar graph of country populations
#
# This code creates a bar chart that visually compares the population of each country.
# Each bar represents a country, and the height of the bar shows the population value.
#
# Step-by-step explanation:
# 1. plt.figure(figsize=(10, 6))
#    - Creates a new figure for the plot with a width of 10 inches and height of 6 inches.
# 2. plt.bar(country_names, populations, color='purple')
#    - Draws a bar for each country, where the height of the bar is the population value. The bars are colored purple.
# 3. plt.title('Population by Country')
#    - Sets the title of the plot.
# 4. plt.xlabel('Country') and plt.ylabel('Population')
#    - Label the x-axis and y-axis.
# 5. plt.xticks(rotation=45)
#    - Rotates the country names on the x-axis by 45 degrees for better readability.
# 6. plt.tight_layout()
#    - Adjusts the layout so labels and titles fit well.
# 7. plt.show()
#    - Displays the plot.

plt.figure(figsize=(10, 6))
plt.bar(country_names, populations, color='purple')
plt.title('Population by Country')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 3. Scatter Plot: Area vs Population

Finally, let's visualize the relationship between country area and population using a scatter plot. Scatter plots are useful for seeing if there's a correlation between two variables.

**Explanation:**
- This code creates a scatter plot where each point represents a country.
- The x-axis is the area, and the y-axis is the population.
- This helps visualize any relationship between area and population (for example, do larger countries tend to have larger populations?).

The following code creates a scatter plot:

```python
# Scatter plot of area vs population
plt.figure(figsize=(8, 6))
plt.scatter(areas, populations, color='teal')
plt.title('Country Area vs Population')
plt.xlabel('Area (sq km)')
plt.ylabel('Population')
plt.grid(True)
plt.show()

# This code creates a scatter plot where:
# - Each point represents a country
# - The x-axis is the area
# - The y-axis is the population
# - This helps visualize any relationship between area and population
```
