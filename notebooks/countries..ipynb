# Web Scraping and Data Extraction: Country Data Example

This notebook demonstrates how to scrape country data from a website using Python libraries such as `requests` and `BeautifulSoup`. The extracted data is then prepared for analysis and visualization. Each step is documented for learning and sharing purposes.

```python
# Import required libraries for web scraping
import requests
from bs4 import BeautifulSoup
```

```python
# Specify the website URL to parse
url = "https://www.scrapethissite.com/pages/simple/"
response = requests.get(url)
```

```python
# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')
# Print the formatted HTML to inspect the structure (for learning/demo purposes)
print(soup.prettify())
```

```python
# Zoom in on the body and section of the page to locate the relevant data
body_section = soup.body.section
body_section  # Display the section containing country data
```

## Extracting Country Data

Now that we've identified the section containing country data, let's extract each country's details for further analysis.

```python
# Find all country divs using their class name
countries = soup.find_all('div', class_='col-md-4 country')
countries  # List of all country elements
```

```python
# Loop through each country and print its details
# This helps verify that we are extracting the correct information
for country in countries:
    name = country.find('h3').text
    population = country.find('span', class_='country-population').text
    area = country.find('span', class_='country-area').text
    print(f"Country: {name}, Population: {population}, Area: {area}")
```

```python
# Extract country details from HTML elements and store them in separate lists.
# This is useful for preparing data for visualization or analysis.

country_names = []         # List to store country names
country_populations = []   # List to store country populations
country_areas = []         # List to store country areas

for country in countries:
    # Extract and clean the country name from the <h3> tag
    name = country.find('h3').text.strip()
    # Extract and clean the population from the <span class='country-population'> tag
    population = country.find('span', class_='country-population').text.strip()
    # Extract and clean the area from the <span class='country-area'> tag
    area = country.find('span', class_='country-area').text.strip()
    
    # Add the extracted values to their respective lists
    country_names.append(name)
    country_populations.append(population)
    country_areas.append(area)

# The three lists now contain the names, populations, and areas of all countries found.
country_names, country_populations, country_areas
```
